```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
labels = pd.DataFrame(iris.target)
labels.columns=['labels']
data = pd.DataFrame(iris.data)
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
data = pd.concat([data,labels],axis=1)

data.head()
```

```python
iris
```

```python
iris.target
```

```python
print(type(iris.target))
```

```python
feature = data[ ['Sepal length','Sepal width','Petal length','Petal width']]
feature.head()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot  as plt
import seaborn as sns
%matplotlib inline

# create model and prediction
model = DBSCAN(eps=0.5,min_samples=5)
predict = pd.DataFrame(model.fit_predict(feature))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([feature,predict],axis=1)

print(r)

# for label in r['predict']:
#     print(label)
```

# DBSCAN 결과 시각화

```python
#pairplot with Seaborn
g = sns.PairGrid(data=r, hue='predict', size=3)
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
g = g.add_legend()
```

# Random data를 KMeans, DBSCAN, HDBSCAN 방식에 따라 군집 분석 실시

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.cluster as cluster
import time
%matplotlib inline
sns.set_context('poster')  # This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style.
sns.set_color_codes()  # Change how matplotlib color shorthands are interpreted.
plot_kwds = {'alpha' : 0.25, 's' : 80, 'linewidths':0}
```

```python
data = np.load('./clusterable_data.npy')
```

```python
data.shape
```

```python
plt.scatter(data.T[0], data.T[1], c='b', **plot_kwds)
frame = plt.gca()  # Get the current Axes instance on the current figure matching the given keyword args, or create one.
frame.axes.get_xaxis().set_visible(False)
frame.axes.get_yaxis().set_visible(False)
```

```python
def plot_clusters(data, algorithm, args, kwds):
    start_time = time.time()
    labels = algorithm(*args, **kwds).fit_predict(data)   # 해당 알고리즘을 통해 주어진 인자를 기반으로 예측 (KMeans, DBSCAN, HDBSCAN)
    end_time = time.time()
    palette = sns.color_palette('deep', np.unique(labels).max() + 1)  # 팔레트의 종류와, 색 종류의 수를 지정
    colors = [palette[x] if x >= 0 else (0.0, 0.0, 0.0) for x in labels]
    plt.scatter(data.T[0], data.T[1], c=colors, **plot_kwds)
    
    frame = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)
    plt.title('Clusters found by {}'.format(str(algorithm.__name__)), fontsize=24)
    plt.text(-0.5, 0.7, 'Clustering took {:.2f} s'.format(end_time - start_time), fontsize=14)
```

```python
plot_clusters(data, cluster.KMeans, (), {'n_clusters':3})
```

```python
plot_clusters(data, cluster.KMeans, (), {'n_clusters':4})
```

```python
plot_clusters(data, cluster.KMeans, (), {'n_clusters':5})
```

```python
plot_clusters(data, cluster.KMeans, (), {'n_clusters':6})
```

```python
# How to find optimal eps
# https://towardsdatascience.com/machine-learning-clustering-dbscan-determine-the-optimal-value-for-epsilon-eps-python-example-3100091cfbc
plot_clusters(data, cluster.DBSCAN, (), {'eps':0.020})
```

```python
plot_clusters(data, cluster.DBSCAN, (), {'eps':0.03})
```

```python
plot_clusters(data, cluster.DBSCAN, (), {'eps':0.010})
```

# DBSCAN 최적의 Eps 값 찾기 (+ KNN 최적의 K값 찾기)

### 1. Cross Validation을 통한 최적의 K값 찾기

```python
from sklearn.model_selection import cross_val_score
from sklearn import neighbors, datasets
```

```python
iris = datasets.load_iris()
labels = pd.DataFrame(iris.target)
labels.columns=['labels']
data = pd.DataFrame(iris.data)
data.columns=['Sepal length','Sepal width','Petal length','Petal width']
new_data = pd.concat([data,labels],axis=1)

new_data.head()
```

```python
# CV 진행
k_range = range(1, 100)
k_scores = []

for k in k_range:
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, data, new_data['labels'], cv = 10, scoring='accuracy')
    k_scores.append(scores.mean())

print(k_scores.index(max(k_scores)))
```

```python
plt.figure(figsize=(15, 10))
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-validated accuracy')
plt.show()
```

### 최적의 K는 12이나, tie problem을 방지하기 위하여 K는 전후값인 11 또는 13을 선택
### K = 13 으로하여, 최적의 eps 찾기

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
```

```python
neigh = NearestNeighbors(n_neighbors=13)   # KNN 모델 생성
nbrs = neigh.fit(data)   # KNN 모델 적합
distances, indices = nbrs.kneighbors(data)   # 거리 및 인덱스 값 연산
```
